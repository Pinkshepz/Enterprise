## Factorial

```python
# input request
number = int(input('Enter an integer: '))

# factorial function
def factorial(num):
    """Calculate factorial of given number"""
    
    # check if num is non-negative integer
    if num < 0:
        return TypeError('Input must be non-negative integer')
    # base case if number = 0
    if num == 0:
        return 1
    # recursively multiply n * n-1 * n-2 * ... until 1
    return num * factorial(num - 1)

# print output
print(factorial(number))
```

## Even sum

```python
# input request
number = int(input('Enter an integer: '))

# even sum function
def even_sum(num):
    """Sum all even numbers from 0 to n"""
    
    # check if num is non-negative integer
    if num < 0:
        return TypeError('Input must be non-negative integer')
    
    # calculate sum of even number series
    half = int(num / 2)
    return int(2 * ((half + 1) * half) * 0.5)

# print output
print(even_sum(number))
```

```python
matches = int(input('Enter number of matches: '))

# function to run the game
def game_engine(matches):
    # declare variables
    score = {'player': 0, 'bot': 0} # {player: score}
    toss_choices = ['rock', 'scissor', 'paper'] # toss choices
    
    print('GAME START')
    print('type rock, scissor or paper')
    
    for match in range(1, matches + 1):
        print('-----MATCH %d-----' % match)
        
        # get player toss
        while True:
            player_toss = input('Your toss: ')
            if player_toss in toss_choices:
                break
            else: 
                print(f'Your toss must be {toss_choices}')
        
        # get bot toss
        bot_toss = toss_choices[npr.randint(3)]
        print('Bot  toss: %s' % bot_toss)
        
        # calculate score
        if (player_toss, bot_toss) in [('rock', 'scissor'), ('scissor', 'paper'), ('paper', 'rock')]:
            print('You win the toss')
            score['player'] += 1
        elif (bot_toss, player_toss) in [('rock', 'scissor'), ('scissor', 'paper'), ('paper', 'rock')]:
            print('Bot win the toss')
            score['bot'] += 1
        else:
            print('Tie')
        
        # print in-game score
        print('Player score: %d     Bot score: %d' % (score['player'], score['bot']))
    
    # print game summary
    print('-----SUMMARY-----')
    if score['player'] > score['bot']: print('You win the game')
    elif score['player'] < score['bot']: print('You lose the game')
    else: print('No winner in the game')
    print('Total match   %d' % matches)
    print('Player score  %d     Bot score  %d' % (score['player'], score['bot']))
    
    return 'GAME END'

game_engine(matches)
```
