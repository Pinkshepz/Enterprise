```python
# Imports
import numpy as np
import pandas as pd
```

## **Q1**

```python
# Define data dictionary
DATA_FRUIT = {
    "fruit": ["Apple", "Orange", "Banana", "Kiwi", "Cherry"],
    "year": [2010, 2005, 2005, 2010, 2005],
    "kg": [30, 100, 50, 20, 10]
}

df_fruit = pd.DataFrame(DATA_FRUIT, index=np.arange(1, 6, 1))

df_fruit
```

```python
# 1.1 Arrange the DataFrame's columns as ["year", "fruit", "kg"]
df_fruit_arr = pd.DataFrame(DATA_FRUIT, columns=["year", "fruit", "kg"], index=np.arange(1, 6, 1))

df_fruit_arr
```

```python
# 1.2 Access the DataFrame with loc[2] and iloc[0]
print(df_fruit.loc[2], df_fruit.iloc[0])
```

```python
# 1.3 Create the new column “price” as shown in (B)
df_fruit["price"] = [1000, 250, 400, 700, 2500]

df_fruit
```

```python
# 1.4 Create the new column “total” = “kg” * “price”
df_fruit["total"] = df_fruit["kg"] * df_fruit["price"]

df_fruit
```

```python
# 1.5 Create the new column “isOrange” (Boolean)
df_fruit["isOrange"] = df_fruit["fruit"] == "Orange"

df_fruit
```

```python
# 1.6 Delete the column “year”
del df_fruit["year"]

df_fruit
```

## **Handling Missing Data**

```python
df = pd.DataFrame(np.random.standard_normal((7, 3)))

df.iloc[:4, 1] = np.nan
df.iloc[:2, 2] = np.nan

df
```

```python
df.fillna({1: 0, 2:np.inf}, limit=1)
```

```python
df = pd.DataFrame(np.random.standard_normal((7, 3)))

df[df.abs() > 1] = np.sign(df) * 500
df
```

```python
df = pd.DataFrame(np.random.standard_normal((10, 10)))

sampler = np.random.permutation(10)

df.take(sampler).sample(n=15, replace=True)
```
